```python
import numpy as np 
import pandas as pd
```

```python
missing_data = { 
                'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Hannah'],  
    'Age': [np.nan, 27, 22, np.nan, 29, 31, np.nan, 28],  
    'City': [np.nan, 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego'], 
    'Score': [85.5, 90.0, np.nan, 88.0, np.nan, 80.0, 87.5, np.nan]
}
df = pd.DataFrame(missing_data)
```

```python
print(df.isnull())
```

```python
df_dropped = df.dropna()
print(df_dropped)
```

```python
fill_age = df['Age'].mean()
df_fill = df.fillna(value={'Age': fill_age, 'Score':10})# understandable 
print(df_fill)
```

```python

duplicate_data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Laptop', 'Monitor', 'Webcam', 'Keyboard', 'Mouse'],
    'OrderID': [101, 102, 103, 101, 104, 105, 103, 106],
    'Category': ['Electronics', 'Accessories', 'Accessories', 'Electronics', 'Electronics', 'Accessories', 'Accessories', 'Accessories'],
    'Price': [1200, 25, 75, 1200, 300, 50, 75, 25]
}
df_dups = pd.DataFrame(duplicate_data)
print(df_dups.duplicated())
```

```python
df_no_dups = df_dups.drop_duplicates()
print(df_no_dups)
```
